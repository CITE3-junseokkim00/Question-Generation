```python
from datasets import load_dataset
```

```python
dataset = load_dataset("squad_kor_v1")
```

```python
import pandas as pd

dataset['train']
```

```python
df_train = pd.DataFrame(dataset['train'])
```

```python
df_train
```

```python
cnt=0
for i in range(len(df_train['answers'])):
    if len(df_train['answers'].iloc[i]['text']) != 1:
        cnt+=1
        continue
    df_train['answers'].iloc[i] = df_train['answers'].iloc[i]['text'][0]
print(f'number of rows which contains more than 1 answers: {cnt}')
```

```python
df_train = df_train[['context', 'question', 'answers']]
```

```python
df_train.to_csv('./data/korquad_1.0/train.csv')
```
